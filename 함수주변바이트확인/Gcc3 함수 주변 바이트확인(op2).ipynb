## 3. 함수의 갯수, 바이트 갯수의 빈공간 np.array 생성 

```python
x = np.full((gcc3_2_32['label'].value_counts()[1],65000), None, dtype='object')
x.shape
x.dtype
```

## 4. 함수 빈공간에 함수별로 한줄씩 값 대입

```python
data = gcc3_2_32.values

row = -1
col = 0
col_list=[]

# data => [55,1], [89,1], [e5,1] ....
for i in data:
    # i[0] = 55, 89, e5... /  i[1] = 1, 0, 0, ....
    if i[1]==1:
        col_list.append(col) # col_list의 각 함수 길이들 넣는다.
        
        row+=1
        col=0
        x[row][col] = i[0]
#        print(row,col, end='/') # 오류났을 때 풀어보면 몇번째 행에서 값을 초과하는지 알 수 있음.
    else:
        x[row][col+1] = i[0]
        #if row >=1385:
        #    print(row,col, end=' ')
        col+=1
    
print('대입 완료')    
```

```python
max(col_list)
```

## 5. 최적의 x 찾기

```python
# 예시
x[:, :6]
#print(len(col_list),'\n')

# 최소 3gram ~ 7gram 까지확인
for j in range(3, 7):
    tuple_ls = []
    for k in x[:, :j]:
        tuple_ls.append(tuple(k))

    set_ls = set(tuple_ls)
    print(j, 'gram일 때 ', len(set_ls))
```
